```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = r'C:\Users\91702\Downloads\tested.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Fare'].fillna(df['Fare'].mean(),inplace=True)
df['Cabin'].fillna('Unknown',inplace=True)
```

```python
df.isnull().sum()
```

```python
sx=df['Sex'].value_counts().reset_index()
a=['male','female']
color=['blue','green']
plt.pie(sx['Sex'],labels=a,autopct='%1.2f%%',colors=color)
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.catplot(x='Pclass', hue='Sex', col='Survived', data=df, kind='count', height=4, aspect=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
